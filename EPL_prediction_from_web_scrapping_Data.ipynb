```python
import pandas as pd


new_df['Count_Home_Win'] = new_df['result'].apply(lambda x: 1 if x == 'W' else 0)
new_df['Count_Draw'] = new_df['result'].apply(lambda x: 1 if x == "D" else 0)
new_df['Count_Away_Win'] = new_df['result'].apply(lambda x: 1 if x == 'L' else 0)


home_team_counts = new_df.groupby('team')[['Count_Home_Win', 'Count_Away_Win', 'Count_Draw']].sum().reset_index()
away_team_counts = new_df.groupby('opponent')[['Count_Home_Win', 'Count_Away_Win', 'Count_Draw']].sum().reset_index()

print("Home Team Counts:")
print(home_team_counts)

print("\nAway Team Counts:")
print(away_team_counts)
```

```python
teams = pd.merge(home_team_counts, away_team_counts, how='outer', left_on='team', right_on='opponent', suffixes=('_Home', '_Away'))

teams['Win'] = home_team_counts['Count_Home_Win'] + away_team_counts['Count_Away_Win']
teams['Loss'] = home_team_counts['Count_Away_Win'] + away_team_counts['Count_Home_Win']
teams['Draw'] = home_team_counts['Count_Draw'] + away_team_counts['Count_Draw']
teams['Total_points'] = (teams['Win']*3) + teams['Draw']
teams['MP'] = teams['Win']+teams['Loss']+teams['Draw']

teams = teams.dropna(subset=['MP'])

teams['MP'] = teams['MP'].astype('int')
teams['Win'] = teams['Win'].astype('int')
teams['Draw'] = teams['Draw'].astype('int')
teams['Loss'] = teams['Loss'].astype('int')
teams['Total_points'] = teams['Total_points'].astype('int')



print("Points Table:")
teams_sorted = teams[['team','MP' ,'Win', 'Draw', 'Loss', 'Total_points']].sort_values(by='Total_points', ascending= False).reset_index(drop= True)
print(teams_sorted)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(teams_sorted['team'], teams_sorted['Total_points'], color='skyblue')

plt.title('Total Points by Team')
plt.xlabel('Team')
plt.ylabel('Total Points')

plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```
